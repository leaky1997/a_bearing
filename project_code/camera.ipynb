```python
from matplotlib.font_manager import FontProperties
import seaborn as sns
myfont = FontProperties(fname=r'/home/user/data/a_bearing/project_code/SIMHEI.TTF', size=14)
# sns.set(font=myfont.get_name())
from pylab import mpl
# 
mpl.rcParams["font.sans-serif"] = ["SimHei"]

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(14, 10))

# Define nodes positions
nodes = {
    '': (0.5, 1),
    '1. ': (0.2, 0.85),
    '2. ': (0.5, 0.85),
    '3. ': (0.8, 0.85),
    '': (0.5, 0.7),
    '1.1 ': (0.1, 0.55),
    '1.2 ': (0.3, 0.55),
    '1.3 ': (0.5, 0.55),
    '1.4 ': (0.7, 0.55),
    '': (0.5, 0.4),
    '': (0.5, 0.25),
    '': (0.5, 0.1),
    '': (0.3, 0.0),
    '': (0.7, 0.0)
}

# Draw nodes
for node, (x, y) in nodes.items():
    ax.text(x, y, node, ha='center', va='center', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=1), fontproperties=myfont)

# Draw edges
edges = [
    ('', '1. '),
    ('', '2. '),
    ('', '3. '),
    ('', '1.1 '),
    ('', '1.2 '),
    ('', '1.3 '),
    ('', '1.4 '),
    ('', ''),
    ('', ''),
    ('', '')
]

for edge in edges:
    start, end = edge
    ax.annotate("",
                xy=nodes[end], xycoords='data',
                xytext=nodes[start], textcoords='data',
                arrowprops=dict(arrowstyle="->", connectionstyle="arc3",
                                edgecolor='black', facecolor='black'),
                )

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.axis('off')
plt.show()
```
